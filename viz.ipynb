```python
import os.path as p

datasets = ['cells', 'polyp', 'aa']

all_experiments = os.listdir('logs')
all_experiments = [e for e in all_experiments if '_cropped_' in e]

def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)
    plt.setp(bp['fliers'], color=color)

fig, axs = plt.subplots(1, 3, figsize=(13, 4))
axs = axs.flatten()
titles = ['Cells', 'Polyp', 'Aorta']

axs[0].set_ylabel('Dice Score Coefficient')
axs[1].set_xlabel('Model input size')

for dataset, ax, title in zip(datasets, axs, titles):
  experiments = [e for e in all_experiments if dataset in e]
  experiments = natsorted(experiments)
  sizes = [e.split('_')[-1] for e in experiments]
  dscs_cropped_all = []
  dscs_uncropped_all = []
  for experiment in experiments:
    dscs_cropped = np.load(p.join('logs', experiment, 'output_metrics_cropped.npy'))[0]
    dscs_uncropped = np.load(p.join('logs', experiment, 'output_metrics_uncropped.npy'))[0]
    dscs_cropped_all.append(dscs_cropped)
    dscs_uncropped_all.append(dscs_uncropped)

  if ax == axs[2]:
    sizes = sizes[::2]
    dscs_cropped_all = dscs_cropped_all[::2]
    dscs_uncropped_all = dscs_uncropped_all[::2]

  bpl = ax.boxplot(dscs_cropped_all, positions=np.array(range(len(dscs_cropped_all)))*2.0-0.4, sym='.', widths=0.6)
  bpr = ax.boxplot(dscs_uncropped_all, positions=np.array(range(len(dscs_uncropped_all)))*2.0+0.4, sym='.', widths=0.6)
  set_box_color(bpl, 'magenta') # colors are from http://colorbrewer2.org/
  set_box_color(bpr, 'orange')


  # draw temporary red and blue lines and use them to create a legend
  ax.plot([], c='magenta', label='Ours')
  ax.plot([], c='orange', label='U-Net')

  ax.set_title(title)

  ax.set_xticks(range(0, len(sizes) * 2, 2), sizes)
  ax.set_xlim(-2, len(sizes)*2)
  ax.set_ylim(0, 1)
axs[0].legend(loc='lower right')
plt.tight_layout()
plt.savefig('plots/box-plots.png', bbox_inches='tight')
```
